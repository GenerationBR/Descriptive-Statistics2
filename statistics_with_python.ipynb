```python
import pandas as pd # biblioteca de analise de dados
import numpy as np # biblioteca numérica
import matplotlib.pyplot as plt # biblioteca para criar graficos
import seaborn as sns # biblioteca de funções estatisticas
```

```python
# quando usamos o seaborn, que é nossa biblioteca de estatística, podemos usar
# ela na criação de gráficos

# configurar o visual dos gráficos pelo seaborn
sns.set_theme(style='whitegrid')

# vamos carregar os dados para uso
df_shoes = pd.read_csv('/content/shoes_sales_dataset.csv')
```

```python
# bloco de funções

# função 1- ver_dados
def ver_dados():
  # o usuario escolhe quantos dados ele quer ver
  resposta = int(input('Informe quantas linhas de daods você quer ver: '))
  df_shoes.head()
  return df_shoes.head(resposta)

# função 2 - Medidas de tendência
# média - moda - mediana
def medidas_tendencia():
  # para as medidas de tendencia, temos funções prontas para cada uma delas
  # mean() - média
  # median() - mediana
  # mode()[0] - moda
  media_precos = df_shoes['Price_USD'].mean()
  media_qtd_produtos = df_shoes['Units_Sold'].mean()
  media_total_vendas = df_shoes['Revenue_USD'].mean()

  # calculos de moda
  moda_cor_primeiro = df_shoes['Color'].mode()[0] # index aqui é 0 - primeiro que apareceu...
  moda_marca = df_shoes['Brand'].mode()[0]

  # calculo de mediana
  mediana_precos = df_shoes['Price_USD'].median()
  mediana_qtd_produtos = df_shoes['Units_Sold'].median()
  mediana_total_vendas = df_shoes['Revenue_USD'].median()

  return (print(f'A média de preços dos produtos é de {media_precos:.2f} \n'
  f'A quantidade média de produtos vendidos é de {media_qtd_produtos:.2f} \n'
  f'A média do total de vendas é de {media_total_vendas:.2f} \n'
  f'O produto com a maior moda é {moda_cor_primeiro} \n'
  f'A marca com a maior moda é {moda_marca} \n'
  f'A mediana nos preços é de {mediana_precos:.2f} \n'
  f'A mediana de produtos vendidos é de {mediana_qtd_produtos:.2f} \n'
  f'A mediana do total de vendas é de {mediana_total_vendas:.2f}'
  ))

def dispersao_variabilidade():
  plt.figure(figsize=(8,5))

  # para tratarmos o tamanho da img
  # o matplotlib, de maneira simples ou facil, não faz bloxplot
  # mas o seaborn faz
  sns.boxplot(y=df_shoes['Revenue_USD'], color='#72ACD0')
  plt.title('Distribuição da receita')

def histograma():
  plt.figure(figsize=(8,5))
  sns.histplot(df_shoes['Price_USD'], kde=True, color ='#72ACD0', bins=40)

def dispersao():
  # precisamos de dois pontos para analise
  plt.figure(figsize=(8,5))
  # o sns.scatterplot(data=DATAFRAME, x='CAMPO X', y='CAMPO Y')
  sns.scatterplot(data=df_shoes, x='Price_USD', y='Revenue_USD', color = '#72ACD0')

def pizza():
  # 1 - preparar
  dados_pizza = df_shoes['Sales_Channel'].value_counts()
  # le dado por dado e mostra suas quantidades

  # 2 - escolher o layout de cores para os setores(cor de cada fatia)
  cores = sns.color_palette('pastel')

  # 3 - criar o gráfico
  plt.figure(figsize=(14,8))
  # plt.pie(os dados, autopct='casas decimais', startangle = angulo de inicio, color)
  plt.pie(dados_pizza,
          labels=dados_pizza.index, # titulo de cada fatia
          autopct='%1.1f%%', # formato de uma casa decimal
          startangle=140, # angulo de inicio
          colors=cores) # cores
```

```python
dispersao_variabilidade()
```

```python
dispersao()
```

```python
pizza()
```
